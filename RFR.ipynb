```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
ds = pd.read_csv("adult.csv")
ds.replace('?', np.nan, inplace=True)
ds.dropna(inplace=True)
```

```python
x=ds.iloc[:,:-1].values
y=ds.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder

ordinal_categorical_columns = [1, 3, 5, 6, 7, 8, 9, 13]

label_encoders = []
for col_index in ordinal_categorical_columns:
    le = LabelEncoder()
    x[:, col_index] = le.fit_transform(x[:, col_index])
    label_encoders.append(le)
```

```python
from sklearn.preprocessing import LabelEncoder
lb_obj= LabelEncoder() 
y=lb_obj.fit_transform(y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model= RandomForestRegressor(n_estimators=10, random_state=0) 
rf_model.fit(x,y)
```

```python
rf_model.predict([[0.177807, 0.088108,0.173244,1.218035,-0.036491,-0.407250,1.289114,-0.898177,0.394225,0.698654,-0.144749,-0.21852,0.851632,0.292299]]) 
```

```python
y_pred = rf_model.predict(x)
plt.figure(figsize=(10, 6), dpi=120)
plt.scatter(np.arange(len(y)), y, color='red', label='Actual')
plt.scatter(np.arange(len(y)), y_pred, color='blue', label='Predicted')
plt.title('Random Forest Regression Results')
plt.xlabel('Data Points')
plt.ylabel('Target Values')
plt.legend()
plt.savefig('random_forest_regression_results.png', dpi=300, bbox_inches='tight')
plt.show()
```
